```python
# Rename columns to be more clear

purchase_gender_analysis.columns=["Purchase Count", "Average Purchase Price", "Total Purchase Value",
                                  "Average Purchase per Person"]
purchase_gender_analysis
```

```python
# Format the data to be more reader friendly

purchase_gender_analysis["Average Purchase Price"] = purchase_gender_analysis["Average Purchase Price"].map("${:.2f}".format)
purchase_gender_analysis["Total Purchase Value"] = purchase_gender_analysis["Total Purchase Value"].map("${:.2f}".format)
purchase_gender_analysis["Average Purchase per Person"] = purchase_gender_analysis["Average Purchase per Person"].map("${:.2f}".format)
purchase_gender_analysis
```

## Most Popular Items

* Retrieve the Item ID, Item Name, and Item Price columns


* Group by Item ID and Item Name. Perform calculations to obtain purchase count, item price, and total purchase value


* Create a summary data frame to hold the results


* Sort the purchase count column in descending order


* Optional: give the displayed data cleaner formatting


* Display a preview of the summary data frame



```python
# groupby_items = purchase_data.groupby(['Item ID','Item Name', 'Price']).agg({'Price': ['count','sum']})
# groupby_items.head()
# groupby_items = groupby_items.reset_index()
# groupby_items.head()
```

```python
popular_item = purchase_data[["Item ID", "Price"]]
# popular_item

# Do a purchase count to see which items were most popular
popular_item_count = purchase_data.loc[:,["Item ID", "Price"]].groupby("Item ID", as_index=False).count()
popular_item_count = popular_item_count.sort_values("Price", ascending=False)
# popular_item_count

# Merge on the Item ID to show the item name and their price with the item count, and remove duplicates
merge_items = pd.merge(popular_item_count, purchase_data.loc[:,["Item ID", 
                                                                "Item Name", 
                                                                "Price"]],
                       on="Item ID").drop_duplicates()
merge_items.head()
```

```python
# Calculate total purchase values
total_price = purchase_data.loc[:,["Item ID", "Price"]].groupby("Item ID",as_index=False).sum()
total_price.head()
```

```python
# Merge total purchase values with merge_items to show ID, purchase count, Item name, Item Price, total revenue
# Merge on the Item ID

merge_all = pd.merge(merge_items, total_price, on="Item ID")
merge_all.head()
```

```python
# Rename the columns
merge_all = merge_all.rename(columns={"Price_x" : "Purchase Count",
                                      "Price_y": "Item Price",
                                      "Price": "Total Purchase Value"})
merge_all.head()
```

```python
# Reorder the columns to match example

merge_all = merge_all[["Item ID", "Item Name", "Purchase Count", "Item Price", "Total Purchase Value"]]
merge_all.head()
```
